```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# 1. CSV 읽기
df = pd.read_csv('pop60up.csv')

# 2. 위경도를 Point로 변환해서 GeoDataFrame 생성
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['lon'], df['lat']), crs='EPSG:4326')

# 3. 행정구역 경계 불러오기 (읍면동 GeoJSON 등)
# 예시: 'emd.geojson' 파일은 미리 받아놓아야 해
boundary = gpd.read_file('emd.geojson')  # 또는 'TL_SCCO_EMD.shp' 같은 shapefile

# 4. 성남시만 필터링
seongnam_boundary = boundary[boundary['SGG_NM'].str.contains('성남')]  # SGG_NM = 시군구명

# 5. 좌표들이 성남시 경계 안에 있는지 확인
gdf_in_seongnam = gpd.sjoin(gdf, seongnam_boundary, how='inner', predicate='within')

# 6. 결과 저장
gdf_in_seongnam.drop(columns='geometry').to_csv('pop60up_seongnam_filtered.csv', index=False, encoding='utf-8-sig')
```
