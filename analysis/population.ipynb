```python
import pandas as pd
```

```python
#파일 불러오기
population = pd.read_csv('C:/가을/대외활동/공모전/성남시 데이터 분석 아이디어 제안 공모전/data/fake_population.csv', encoding='utf-8') 
population
```

```python
#60세 이상 인구 추출
pop60up = population[['HCODE','X_COORD','Y_COORD','M_6064','M_6569', 'M_7000','W_6064', 'W_6569', 'W_7000',]]
pop60up['Mtotal'] = population[['M_6064','M_6569', 'M_7000']].sum(axis=1)
pop60up['Wtotal'] = population[['W_6064','W_6569', 'W_7000']].sum(axis=1)
pop60up['Total'] = population[['M_6064','M_6569', 'M_7000','W_6064','W_6569', 'W_7000']].sum(axis=1)
pop60up
```

```python
from pyproj import Transformer

# EPSG:5179 → EPSG:4326 (위도/경도) 변환기 정의
transformer = Transformer.from_crs("EPSG:5179", "EPSG:4326", always_xy=True)

# 변환 적용
lon, lat = transformer.transform(pop60up['X_COORD'].values, pop60up['Y_COORD'].values)

# DataFrame에 위도/경도 열 추가
pop60up['lon'] = lon
pop60up['lat'] = lat
```

```python
# 좌표도 평균으로 포함하고 싶을 경우
grouped= pop60up.groupby('HCODE', as_index=False).agg({
    'Mtotal': 'sum',
    'Wtotal': 'sum',
    'Total': 'sum',
    'X_COORD': 'mean',
    'Y_COORD': 'mean',
    'lon': 'mean',
    'lat': 'mean'
})

print(grouped)
```

```python
grouped.to_csv('pop60up_grouped2.csv', index=False, encoding='utf-8-sig')
```
